```python
vacmot = (vac.iloc[:, 12:31] == "yes").astype(int)
```

```python


vacmot_scaled = preprocessing.scale(x)
```

```python
#Principal component analysis

from sklearn.decomposition import PCA
from sklearn import preprocessing

pca = PCA()
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11','pc12','pc13','pc14','pc15','pc16','pc17','pc18','pc19']
pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
#Proportion of Variance (from PC1 to PC11)
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = vacmot.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
#Correlation matrix plot for loadings 
plt.rcParams['figure.figsize'] = (25,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
plt.show()
```

```python
#Scree plot (Elbow test)- PCA
from bioinfokit.visuz import cluster
cluster.screeplot(obj=[pc_list, pca.explained_variance_ratio_],show=True,dim=(10,5))
```

```python
# get PC scores
pca_scores = PCA().fit_transform(x)

# get 2D biplot
cluster.biplot(cscore=pca_scores, loadings=loadings, labels=vacmot.columns.values, var1=round(pca.explained_variance_ratio_[0]*100, 2),
    var2=round(pca.explained_variance_ratio_[1]*100, 2),show=True,dim=(20,12))
```

```python
#Extracting segments

#Using k-means clustering analysis
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(vacmot)
visualizer.show()
```

```python
#K-means clustering 

kmeans = KMeans(n_clusters=2, init='k-means++', random_state=0).fit(x)
vac['cluster_num'] = kmeans.labels_ #adding to df
print (kmeans.labels_) #Label assigned for each data point
print (kmeans.inertia_) #gives within-cluster sum of squares. 
print(kmeans.n_iter_) #number of iterations that k-means algorithm runs to get a minimum within-cluster sum of squares
print(kmeans.cluster_centers_) #Location of the centroids on each cluster. 
```

```python
#To see each cluster size
from collections import Counter
Counter(kmeans.labels_)
```

```python
#Visulazing clusters
sns.scatterplot(data=pf, x="pc1", y="pc2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```
